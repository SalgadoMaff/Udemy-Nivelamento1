```python
import re
class item_Pedido:
  def __init__(self, codigo:int, num_pecas:int, valor_uni:float):
    self.codigo = codigo
    self.num_pecas = num_pecas
    self.valor_uni = valor_uni
  def calcular(self)->float:
    return(self.valor_uni*self.num_pecas)


qtd = input("Digite a quantidade de items diferentes que vai comprar: ")
if(len(qtd)<1 or qtd.isspace()):
  raise TypeError("Digite uma quatidade não nula!")

elif (int(qtd)<=0):
  raise TypeError(f'Digite uma quantidade maior que 0')

else:
  Pedidos = []
  Total = float(0)
  for i in range(0,int(qtd)):
    string = input('Digite as entradas tal qual mostrado no exemplo acima: ')
    separados = string.split(sep=' ')
    if (len(separados) == 3):
      if (separados[0].isdigit and separados[1].isdigit):
        preco=re.findall('\d+?.\d+', separados[2])[0]
        if (len(preco)>0):
          Pedidos.append(item_Pedido(int(separados[0]), int(separados[1]),float(preco)))
          Total += Pedidos[i].calcular()
        else:
          raise Exception("Digite uma entrada válida!!")
      else:
        raise Exception("Digite uma entrada válida!!")
    else:
      raise Exception('Digite uma entrada válida!!')

print('Valor A PAGAR: R$' + '{:.2f}'.format(Total))
```

```python
entrada = input("'a','b' e 'c' conforme o exemplo acima: ")
if(len(entrada)<1 or entrada.isspace()):
  raise TypeError("Digite uma quatidade não nula!")
else:
  separados = entrada.split(sep=' ')
  if(len(separados) == 3):
    a = float(separados[0])
    b = float(separados[1])
    c = float(separados[2])
    delta = pow(b,2)-(4*a*c)
    if( delta < 0 or a == 0):
      print("Impossível calcular")
    else:
      x1 = (-b+pow(delta,(1/2)))/(2*a)
      x2 = (-b-pow(delta,(1/2)))/(2*a)
      print('X1 = {:.5f}'.format(x1) + '\nX2 = {:.5f}'.format(x2))


  else:
    raise Exception("Digite a quantidade certa de argumentos!")
```
